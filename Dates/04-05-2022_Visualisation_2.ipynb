```python
import pandas as pd
import matplotlib.pyplot as plt

gapminder = pd.read_csv('https://raw.githubusercontent.com/rogovich/Data/master/data/gapminderData.csv')
gapminder['continent'] = pd.Categorical(gapminder['continent'])

for year in gapminder.year.unique():
    fig, ax = plt.subplots(figsize=(14, 8))
    gapminder_year = gapminder[gapminder['year'] == year]
    ax.scatter(gapminder_year['gdpPercap'], gapminder_year['lifeExp'],
              s=gapminder_year['pop'] / 100000, alpha=0.7,
              c=gapminder_year['continent'].cat.codes,
              cmap='Set2') # color map
    ax.set_ylim(0, 100)
    ax.set_xlim(100, 100000)
    ax.set_xscale('log')
    ax.set_title(f'   \n     . : {year}')
    ax.set_xlabel('   ')
    ax.set_ylabel('  ')
    x_china = gapminder_year[gapminder_year['country'] == 'China']['gdpPercap']
    y_china = gapminder_year[gapminder_year['country'] == 'China']['lifeExp']
    ax.annotate('', (x_china, y_china))
    fig.savefig(f'{year}.pdf')
#   x,  y,  s=  , c=      , cmap =  
```

```python
crimes = pd.read_csv('https://raw.githubusercontent.com/rogovich/Data/master/data/crimeRatesByState2005.tsv', sep='\t')  #   -  
crimes.head()
```

```python
fig, ax = plt.subplots(figsize=(14,8))
ax.scatter(crimes.murder, crimes.burglary, s=crimes.population/100000, alpha=0.6, c=crimes.motor_vehicle_theft)
proxy = crimes[crimes.burglary >= crimes.burglary.max()]
x_murder = proxy.murder
y_burglary = proxy.burglary
ax.annotate(crimes[crimes.burglary >= crimes.burglary.max()].state.item(), (x_murder, y_burglary))
```

```python
fig, ax = plt.subplots(figsize=(14,8))
ax.scatter(crimes.murder, crimes.burglary, s=crimes.population/30000, alpha=0.5, c=crimes.motor_vehicle_theft)

for state in crimes.state.unique():
    filtered = crimes[crimes.state == state]
    x_murder = filtered.murder
    y_burglary = filtered.burglary
    ax.annotate(state, (x_murder, y_burglary))
```
