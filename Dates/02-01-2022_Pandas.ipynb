```python
def get_temp_cat(temp):
    if temp < 0:
        return 'Freezing'
    elif temp < 15:
        return 'Chilly'
    elif temp < 26:
        return 'Nice'
    elif temp >= 26:
        return 'Hot'
    else:
        return temp
    
bikes['Temp_Cat'] = bikes['Temperature'].apply(get_temp_cat)
bikes.head()
```

```python
import numpy as np

bikes['a'] = np.where((bikes['Holiday'] == True) & (bikes['Seasons'] == 'Winter'), 1, 0)
bikes.head()
```
