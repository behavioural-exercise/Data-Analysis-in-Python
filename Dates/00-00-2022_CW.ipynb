```python
country_results = []
for country in ['AL', 'AT', 'BG', 'HR', 'CY', 'DK', 'EE', 'FI', 'FR', 'DE', 'GR', 'HU',
                'IT', 'LV', 'LT', 'ME', 'PL', 'PT', 'RO', 'RS', 'SK', 'SI',
                'ES', 'SE', 'CZ', 'NL', 'MK', 'TR']:
    country_score = []
    for variable in ten:
        wvsc = wvs[wvs.cntry_AN == country]
        side1 = wvsc[variable].value_counts(normalize=True)[[1,2,3]].sum().round(3)
        side2 = wvsc[variable].value_counts(normalize=True)[[8,9,10]].sum().round(3)
        side3 = wvsc[variable].value_counts(normalize=True)[[4,5,6,7]].sum().round(3)

        if (side1 > 0.15 and side2 > 0.15) and (side1 <=0.50 and side2 <=0.50 and side3 <=0.51):
            variable_score =[]
            for i in wvsc[variable]:
                if i in [5,6]:
                    variable_score.append(0)
                elif i in [4,7]:
                    variable_score.append(0.5)
                elif i in [3,8]:
                    variable_score.append(1)
                elif i in [2,9]:
                    variable_score.append(2)
                elif i in [1,10]:
                    variable_score.append(3)   
            country_score.append(pd.DataFrame(variable_score).mean().item())
            sns.histplot(wvsc[variable], discrete=True)
            plt.show()
            print(f'OK\nПервое крыло поляризации: {side1}\nВторое крыло поляризации: {side2}\nКоличество умеренных: {side3}')
            print(f'\nПоказатели страны {country} {country_score}\n\n')
        else:
            sns.histplot(wvsc[variable], discrete=True)
            plt.show()
            print('NOT OK')
            print(f'Первое крыло поляризации: {side1}\nВторое крыло поляризации: {side2}\nКоличество умеренных: {side3}')
            country_score.append(0)
            print(f'Показатели страны {country} {country_score}\n\n')
    country_results.append(pd.DataFrame(country_score).mean().item())
```
