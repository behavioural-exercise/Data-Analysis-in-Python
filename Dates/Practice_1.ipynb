## Задание 7 (создание новых переменных)

*2.5 балла*

Добавьте новый категориальный признак **bodyType** на основе столбца `BMI`:

BMI Categories:

* Underweight = <18.5

* Normal weight = 18.5–24.9

* Overweight = 25–29.9

* Obesity = BMI of 30 or greater

Признак должен принимать значения **Underweight**, **Normal weight**, **Overweight** и **Obesity**.

```python
# ваш код здесь
def get_type(blood):
    if blood <18.5:
        return 'Underweight'
    elif blood <24.9:
        return 'Normal weight'
    elif blood <29.9:
        return 'Normal weight'
    elif blood >= 30:
        return 'Obesity'
    else:
        return blood
```

```python
data['bodyType'] = data['BMI'].apply(get_type)
data[0:50].sort_values('BMI')
```
