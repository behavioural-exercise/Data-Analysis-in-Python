```python
### country_results_variable = []
country_results = []
for country in ['AL', 'AT', 'BG', 'HR', 'CY', 'DK', 'EE', 'FI', 'FR', 'DE', 'GR', 'HU',
                'IT', 'LV', 'LT', 'ME', 'PL', 'PT', 'RO', 'RS', 'SK', 'SI',
                'ES', 'SE', 'CZ', 'NL', 'MK', 'TR']:
    country_score = []
    for variable in ten:
        wvsc = wvs[wvs.cntry_AN == country]
        side1 = wvsc[variable].value_counts(normalize=True)[[1,2,3]].sum().round(3)
        side2 = wvsc[variable].value_counts(normalize=True)[[8,9,10]].sum().round(3)
        side3 = wvsc[variable].value_counts(normalize=True)[[4,5,6,7]].sum().round(3)

        if (side1 > 0.15 and side2 > 0.15) and (side1 <=0.50 and side2 <=0.50 and side3 <=0.51):
            variable_score =[]
            for i in wvsc[variable]:
                if i in [5,6]:
                    variable_score.append(0)
                elif i in [4,7]:
                    variable_score.append(0.5)
                elif i in [3,8]:
                    variable_score.append(1)
                elif i in [2,9]:
                    variable_score.append(2)
                elif i in [1,10]:
                    variable_score.append(3)
            country_score.append(pd.DataFrame(variable_score).mean().item())
        else:
            country_score.append(0)
    country_results.append(pd.DataFrame(country_score).mean().item())
    country_results_variable.append(country_score)
```

```python
# данные взяты из cmpf.eui.eu/mpm-2021-interactive/. Concentration = market plurality/news media concentration, mpmAll = market plurality/all indicators
# mpmall - более широкое понимание концентрации медиа, при котором учитывается контекст страны относительно ее регулирования медиа и т.д.

data = pd.DataFrame()
data['code'] = ['AL', 'AT', 'BG', 'HR', 'CY', 'DK', 'EE', 'FI', 'FR', 'DE', 'GR', 'HU',
                'IT', 'LV', 'LT', 'ME', 'PL', 'PT', 'RO', 'RS', 'SK', 'SI',
                'ES', 'SE', 'CZ', 'NL', 'MK', 'TR']
data['polar'] = country_results
data['concentration'] = [89,82,96,68,70,70,76,90,75,57,49,84,83,81,77,74,92,85,79,83,83,85,85,96,89,85,56,68]
data['concentration'] = data.concentration/100
data['mpmAll'] = [89,63,82,71,73,51,64,74,58,37,66,82,66,75,70,62,74,60,84,69,79,76,75,62,85,57,64,76]
data['mpmAll'] = data.mpmAll/100
data.head()
```

```python
data['concentration'] = (data.concentration - data.concentration.min()) / (data.concentration.max() - data.concentration.min())
data['polar'] = (data.polar - data.polar.min()) / (data.polar.max() - data.polar.min())
data['mpmAll'] = (data.mpmAll - data.mpmAll.min()) / (data.mpmAll.max() - data.mpmAll.min())
data.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(13, 6))

for variable, axis in zip(['concentration', 'mpmAll'], ax.ravel()):
    axis.plot(data[variable], data.polar, marker='.', linestyle='none')
    axis.margins(0.02)
    a, b = np.polyfit(data[variable], data.polar, 1)
    x = np.array([0, 1])
    y = a*x+b
    axis.plot(x, y)
    axis.grid(True)
```
